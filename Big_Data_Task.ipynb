```python
import pandas as pd
import csv
import os


# Set up file paths
csv_file_path = "MetObjects.csv"
output_dir = "/content"
malformed_records_file_path = os.path.join(output_dir, "malformed_records.csv")
malformed_records_list=[]
# Read the CSV file and handle malformed records
try:
    df = pd.read_csv(csv_file_path)
except pd.errors.ParserError as e:
    # Handle the ParserError and store the malformed records
    malformed_records = e.args[0].split('\n')
    malformed_records_list.append[malformed_records]

    
    # Write the malformed records to a separate file
with open(malformed_records_file_path, 'w') as file:
             for record in  malformed_records_list :
              file.write(record + '\n') 
    
    # Continue reading the remaining valid records
df = pd.read_csv(csv_file_path, error_bad_lines=False, warn_bad_lines=True)



```

#Pre-Processing

###a) Dimensions column

```python
#Show Dim Column to see its  structure
df['Dimensions']
```

```python
#Remove Dims in inches and remain only cm 
df["New_Dimensions"] = df["Dimensions"].str.extract(r"\((.*?)\)" , expand=False)
df["New_Dimensions"]
```

```python
#Remove word cm
df['New_Dimensions'] = df['New_Dimensions'].str.replace('cm','')
df['New_Dimensions']
```

```python
#Function to clean the dim column 
def clean_Dim(dim):
    if isinstance(dim, str):
        dim = dim.strip()  # Remove leading/trailing whitespaces
        dim = dim.replace(' x ', ' ')  
        dim = dim.replace(' H.', ' ')  
        dim = dim.replace(' dwt ', '')  
        dim = dim.replace('Diam. ', '') 

    return dim
```

```python
df['New_Dimensions'] = df['New_Dimensions'].apply(clean_Dim)
```

```python
df['New_Dimensions'].unique().tolist()
```

```python
#function to split the dimension column into three columns
def split_dimensions(row):
    dimensions = str(row).split(" ")
    if len(dimensions) == 3:
        return [dimension.strip() for dimension in dimensions]
    if len(dimensions) == 2:
        return [dimensions[0].strip(), dimensions[1].strip() , 'null']
    elif len(dimensions) == 1:
        return [dimensions[0].strip(), 'null' ,'null']
    else:
        return ['null', 'null','null']
```

```python
#Apply Function on our dataframe
df[['hight_col', 'width_col', 'length_col']] = df["New_Dimensions"].apply(lambda x: pd.Series(split_dimensions(x)))

```

###b)Country Column

```python
#Check uniqe values on country column
df['Country'].unique().tolist()

```

```python
#Function to clean the country column 
def clean_country(country):
    if isinstance(country, str):
        country = country.strip()  # Remove leading/trailing whitespaces
        country = country.title()  # Capitalize the first letter of each word
        country = country.replace(' Or ', ' , ')  # Replace ' or ' with ','
        country = country.replace('|', ' , ')  # Replace ' | ' with ','
        country = country.replace('Present-Day ', '')  # Remove 'Present-Day ' prefix
        country = country.replace('Probably ', '')  # Remove 'Probably ' prefix
        country = country.replace('Possibly ', '')  # Remove 'Possibly ' prefix
        country = country.replace('Unknown Country', 'null')  # Replace 'Unknown Country' with NaN
        country = country.replace('?', '')  # Remove '?' if present
        country = country.replace('()', '')  # Remove '()' if present

    return country
```

```python
#Apply Function one our DF
df['New_Country'] = df['Country'].apply(clean_country)
```

I will choose Relational Database:








Useful: Storing data in a relational database allows for a more structured and organised approach of storing and managing data. It enables efficient querying, indexing, and the maintenance of relationships between tables.

Advantages:

1)Data Integrity: Through limitations such as primary keys, foreign keys, and data types, relational databases ensure data accuracy and consistency.

2)Scalability: Relational databases are built to manage enormous datasets and allow several users to access them at the same time.


3)SQL (Structured Query Language) allows for complicated searches, aggregations, and join operations on data, making it simple to extract significant insights.


Disadvantages:

1)Complexity: Setting up and managing a relational database involves technical knowledge as well as administrative abilities.

2)Cost: Depending on the database system, there may be licensing or infrastructure costs associated with using a relational database.

Without implementing any code, explain what approaches you would take to optimize the performance of a data pipeline.

1) Use Distributed Computing

Distributed computing is one of the most effective methods for accelerating your data flow. This comprises breaking down data processing into smaller components that can be handled simultaneously by many machines. This allows you to process more data in less time by drastically lowering processing time and increasing throughput.

2) Choose the Correct Data Storage

The storage technology you choose may have a considerable impact on the performance of your pipeline. Data size, processing requirements, and budget are all important considerations. If you're dealing with large amounts of structured data, a relational database can be a good option. If you're working with unstructured data, a NoSQL database may be more appropriate. As a result, if you need to strike a balance between performance and cost, a cloud-based storage solution like Amazon S3 may be the ideal alternative.

3) Optimize Data Compression

Data compression is also required to increase the efficiency of your data pipeline. By compressing your data, you can reduce its size and thus its storage and processing costs. This can help save processing time when only a little amount of data needs to be carried through the pipeline. However, compression must be calibrated with data quality and processing requirements, since too much compression may have a negative impact on performance.Price and performance must be balanced.

4) Data Quality and Governance

Data quality and governance are essential components of an effective data pipeline. Data quality refers to the accuracy, completeness, and consistency of data being processed, and is critical to ensuring that data pipeline outputs are usable and valuable.
